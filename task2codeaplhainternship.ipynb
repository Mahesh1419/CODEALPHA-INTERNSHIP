```python
# Plotting using Seaborn Relplot
sns.relplot(data=my_dataset, x=' Date', y=' Estimated Unemployment Rate (%)', kind='line', height=5, aspect=3)

plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate Over Time')
plt.show()

```

```python
covid_beforeone = my_dataset[my_dataset['Date'] < '2020-03-31']
covid_afterone = my_dataset[my_dataset['Date'] >= '2020-06-30']

plt.figure(figsize=(18, 14))
sns.lineplot(data=covid_beforeone, x='Date', y='Estimated Unemployment Rate (%)', label='Pre COVID-19')
sns.lineplot(data=covid_afterone, x='Date', y='Estimated Unemployment Rate (%)', label='Post COVID-19')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Before and After COVID-19')
plt.legend()
plt.show()

```
